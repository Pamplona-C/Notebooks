```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split #Essa função é usada para dividir o conjunto de dados em duas (ou mais) partes: uma para treinamento e outra para teste
from sklearn.linear_model import LogisticRegression  #algoritmo de classificação usado para prever categorias binárias (ou múltiplas, com algumas modificações)
from sklearn.metrics import accuracy_score #mede a acurácia do modelo, que é a proporção de previsões corretas feitas pelo modelo em comparação com o total de previsões.
from sklearn.metrics import mean_squared_error #calcula a média dos quadrados das diferenças entre os valores previstos e os valores reais. Quanto menor o valor, melhor o modelo.
from sklearn.metrics import r2_score #mede o quão bem os valores previstos pelo modelo se ajustam aos dados reais.
from sklearn.metrics import confusion_matrix #mostra a relação entre as previsões feitas pelo modelo e os rótulos reais dos dados
```

```python
#Importando o arquivo
df = pd.read_csv('creditcard.csv')
```

```python
#verifica Data frame
df.head()
```

```python
#verifica tipos
df.dtypes
```

```python
#verificando se existe valores nulos
df.isnull().sum()
```

```python
#observado que existe valores nulos, agora importa que ajustemos isso, trocando os valores nulos por 0
df = df.fillna(0)
df.isnull().sum()
```

```python
#observado que a coluna class esta do tipo float, troquei para o tipo int
df['Class'] = df['Class'].astype(int)
df.dtypes
```

```python
df.isnull().sum()
```

```python
#verifica os tipos dos dados não fraude
df_nao_fraude = df.Amount[df.Class == 0]

df_nao_fraude.describe()
```

Observando as transações não fraude, temos que dentro do data set temos 45504 transações não fraudulentas.
A média das transações foi 90,00 dólares,Um desvio padrão de 240, o mínimo foi 0,00, o valor máximo foi de 7.878,00 e temos os percentis onde em 25% dos casos os valores foram de 7,58 ou menos, em 50% dos casos foram valores 24,99 e 75% dos casos 82,00 ou menos

```python
#verifica os tipos de dados de fraude
df_fraude = df.Amount[df.Class == 1] #relacionado a coluna amount com a coluna class onde class é igual a 1
#Vai retornar uma coluna com os valores de amount onde a coluna class é igual a 1

df_fraude.describe()
```

Observando os dados das transações fraudulentas temos que 142 transações no data set são fraudulentas, com uma média de valores de 97 dólares, um desvio padrão de 233,00 dólares, oque indica uma variabilidade alta nos valores das transações.
